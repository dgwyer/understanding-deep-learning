```python
import numpy as np
import matplotlib.pyplot as plt
```

# Linear Regression!
This notebook is meant to explore linear regressions and explore how to best fit a line to some data points (here I used 3). This notebook follows Chapter 2 of the Understanding Deep Learning book by Simon J.D. Prince.

```python
# Prepare 3 data points
data = np.array([[2, 4, 7], [4, 6, 5]])
```

```python
# Show initial graph
x = np.arange(0, 10, 0.01)

# Set up an arbitraty phi1 and phi0
phi1 = 1
phi0 = 0

ax = plt.subplot()
y = x * phi1 + phi0
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)
```

```python
# evaluate loss
loss = (((data[0] * phi1 + phi0) - data[1])**2).sum()
loss
```

```python
# calculate gradients
dphi0 = 2 * (phi1 * data[0] + phi0 - data[1]).sum()
dphi1 = 2 * (data[0] * (phi1 * data[0] + phi0 - data[1])).sum()
dphi0, dphi1
```

```python
# 1 training iteration
lr = 0.01
phi1 -= lr * dphi1
phi0 -= lr * dphi0
```

```python
# loss is down! yay!
loss = (((data[0] * phi1 + phi0) - data[1])**2).sum()
loss, phi1, phi0
```

```python
# curve budged a bit
ax = plt.subplot()
y = x * phi1 + phi0
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)
```

# Exercise 2.2

Here I computed the ideal values of phi1 and phi0 for the best fit line, then plotted it to see how it looks.

```python
# jump straight to lowest loss (Exercise 2.2)
real_phi1 = (-(data[1] * data[0]).sum() / data[0].sum() + len(data[0]) ** -1 * data[1].sum()) * (len(data[0]) ** -1 * data[0].sum() - (data[0] ** 2).sum()/data[0].sum()) ** -1
real_phi0 = -(real_phi1 * data[0] **2 - data[1] * data[0]).sum() / data[0].sum()
```

```python
real_phi1, real_phi0
```

```python
loss = ((real_phi1*data[0] + real_phi0 - data[1])**2).sum()
loss
```

```python
# Show the best fit line
ax = plt.subplot()
y = x * real_phi1 + real_phi0
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)
```

# Gradient descent
Here I used gradient descent to check whether the line would fit similarly to that of the one found with the closed form solution

```python
phi1, phi0
```

```python
# Fitting the line with gradient descent
for i in range(500):
    dphi0 = 2 * (phi1 * data[0] + phi0 - data[1]).sum()
    dphi1 = 2 * (data[0] * (phi1 * data[0] + phi0 - data[1])).sum()
    phi1 -= lr * dphi1
    phi0 -= lr * dphi0
    loss = (((data[0] * phi1 + phi0) - data[1])**2).sum()
    if i % 10 == 0: print(loss)

ax = plt.subplot()
y = x * phi1 + phi0
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)

```

# Exercise 3
Here I tried to fit a line with gradient descent and see the resulting line (I really dont want to derive the whole thing again got this one)

```python
# Exercise 2.3 thingies
phi1 = 1
phi0 = 0

ax = plt.subplot()
y= (x-phi0)/phi1
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)
```

```python
loss = (((data[0] - phi0)/phi1 - data[1])**2).sum()
loss
```

```python
dphi1 = -phi1**-2 * 2 * (((data[0] - phi0)/phi1 - data[1]) * (data[0] - phi0)).sum()
dphi0 = -2 * phi1 ** -1 * ((data[0]-phi0)/phi1 - data[1]).sum()

dphi1, dphi0
```

```python
lr = 0.01
phi1 -= lr * dphi1
phi0 -= lr * dphi0

loss = (((data[0] - phi0)/phi1 - data[1])**2).sum()
loss, dphi1, dphi0
```

```python
# Fitting line for the inverse function (fits a lot slower), 5000 iterations and still not lowest loss
for i in range(5000):
    dphi1 = -(phi1**-2) * 2 * (((data[0] - phi0)/phi1 - data[1]) * (data[0] - phi0)).sum()
    dphi0 = -2/phi1 * ((data[0]-phi0)/phi1 - data[1]).sum()
    phi1 -= lr * dphi1
    phi0 -= lr * dphi0
    loss = (((data[0] - phi0)/phi1 - data[1])**2).sum()
    if i % 10 == 0: print(loss)

ax = plt.subplot()
y= (x-phi0)/phi1
ax.scatter(data[0], data[1])
ax.plot(x, y)
ax.set(xlim=(0,10), ylim=(0, 10))
plt.show(ax)
```
